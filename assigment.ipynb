```python
import pandas as pds
import requests as rq
from bs4 import BeautifulSoup
import numpy as np
```

```python
url="https://www.flipkart.com/search?q=iphone&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_1_1_na_na_ps&otracker1=AS_Query_OrganicAutoSuggest_1_1_na_na_ps&as-pos=1&as-type=HISTORY&suggestionId=iphone&requestId=ea6d2f90-a52f-40e9-ad04-ef757b2497b6"
```

```python
response=rq.get(url)
response
```

```python
html=BeautifulSoup(response.text)
html
```

```python
Amount=html.select("._1_WHN1")
Amount
```

```python
Name=html.select("._4rR01T")
Name
```

```python
Details=pds.DataFrame(Name,columns=["Phone Name"])
Details["Amount"]=Amount
Details
```
