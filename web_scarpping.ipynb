```python
from bs4 import BeautifulSoup
import pandas as ps
import numpy as np
```

```python
import requests
```

```python
url="https://www.imdb.com/list/ls091520106/"
```

```python
response=requests.get(url)
```

```python
print(response)
```

```python
html=BeautifulSoup(response.text)
html
```

```python
title=html.select(".lister-item-header a")
```

```python
title
```

```python
runtime=html.select(".runtime")
```

```python
runtime
```

```python
detail=ps.DataFrame(title,columns=['Movie name'])
detail['Runtime']=runtime
```
